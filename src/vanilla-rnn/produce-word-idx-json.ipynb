```python
# STEP ONE: Read from all the data
dictionary = {
    "idx_to_word": [],
    "word_to_index": {}
}

artist_dict = {
    "idx_to_artist": [],
    "artist_to_index": {}
}
```

```python
import csv
import re

unique_words = []
artist_list = []
with open("./data/big-kaggle-dataset.csv") as dataset:
  reader = csv.reader(dataset)
  line = 1
  for row in reader:
    if line > 1:
      # artist
      artist = row[0]
      artist_list.append(artist)
      artist_list = list(set(artist_list))
      # lyrics
      lyrics_list = row[1].lower() #.split(" ")
      #lyrics_list = re.sub(r'[!.?)(,\[\]\"\':|;*\{\}]', '', lyrics_list)
      lyrics_list = re.sub(r'[-]', ' ', lyrics_list)
      lyrics_list = re.sub(r'[^\s+a-z+]+', '', lyrics_list).split(" ")
      lyrics_list = set(lyrics_list)
      unique_words += list(lyrics_list)
    line += 1
```

```python
print(len(unique_words))
unique_words = list(set(unique_words))
print(len(unique_words))
print(unique_words[:50])
print(len(artist_list))
```

```python
# unique_words will correspond to the first key,value pair
unique_words = ["<NULL>", "<UNK>", "<START>", "<END>"] + unique_words
dictionary["idx_to_word"] = unique_words
artist_dict["idx_to_artist"] = artist_list
```

```python
# go through each word here, and get the key,val pair
word_to_idx_dict = {}
for i, word in enumerate(unique_words):
  word_to_idx_dict[word] = i

artist_to_idx_dict = {}
for i, artist in enumerate(artist_list):
  artist_to_idx_dict[artist] = i
```

```python
dictionary["word_to_index"] = word_to_idx_dict
artist_dict["artist_to_index"] = artist_to_idx_dict
```

```python
import json

json_object = json.dumps(dictionary, indent=4)
artist_json_object = json.dumps(artist_dict, indent=4)

with open("word-to-idx.json", "w") as outfile:
  outfile.write(json_object)

with open("artist-to-idx.json", "w") as artistfile:
  artistfile.write(artist_json_object)
```
