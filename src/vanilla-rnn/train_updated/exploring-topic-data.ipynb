```python
# open the updated csv files
import csv
topic_words_list = []
topic_no_to_words = []
sentence_dict = {}
with open("./train_updated/big-lda-train-40.csv") as ogfile:
  reader = csv.reader(ogfile, delimiter=",")
  line = 1
  for row in reader:
    if line > 1:
      topic_str = row[1]
      #topic_no_to_words.append(topic_str)
      sentence_dict[topic_str.split(": ")[0]] = topic_str.split(": ")[1]
      topic_words_list.append(topic_str.split(": ")[1])
    line += 1
    
topic_words_list = set(topic_words_list)
topic_no_to_words = set(topic_no_to_words)
print(len(topic_words_list))
print(len(topic_no_to_words))
print(topic_no_to_words)
```

```python
# write the topic # to word list json
import json
json_object = json.dumps(sentence_dict)

with open("./train_updated/lda-40-number-to-list.json", "w") as outfile:
    outfile.write(json_object)
```

```python
# WE'RE GONNA BE USING GLOVE FROM STANFORD
from gensim.models import KeyedVectors
from gensim.scripts.glove2word2vec import glove2word2vec
from gensim.test.utils import datapath, get_tmpfile
import numpy as np


glove_file = './glove.6B.50d.txt'
tmp_file = get_tmpfile("lda-40-word2vec.txt")
_ = glove2word2vec(glove_file, tmp_file)
model = KeyedVectors.load_word2vec_format(tmp_file)
```

```python
embedded_sentences = {}
for topic, sentence in sentence_dict.items():
  embedded_sent = []
  #print(topic)
  #print(sentence)
  for word in sentence.split(" "):
    #print(word)
    try:
      embed_word = model.get_vector(word)
    except:
      print("error lol")
      embedded_sent.append(np.zeros((50,)))
      continue
    #print(embed_word)
    embedded_sent.append(embed_word)
    #print(embedded_sent)
    #embedded_sent.append(model.get_vector(word))
  embedded_sentences[int(topic)] = embedded_sent
  #embedded_sentences.append(embedded_sent)

#Show a word embedding
print(embedded_sentences)
```

```python
print(len(embedded_sentences))
print(embedded_sentences[0])
```

```python
for topic, embed_list in embedded_sentences.items():
  embedded_sentences[topic] = np.concatenate(embedded_sentences[topic], axis=0).tolist()
```

```python
# write the topicword-to-idx.json file 
json_object = json.dumps(embedded_sentences)

with open("./train_updated/lda-40-topic-embeddings.json", "w") as outfile:
    outfile.write(json_object)
```
