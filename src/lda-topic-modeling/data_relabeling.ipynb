```python
# mount google drive
from google.colab import drive
import os
import gensim
from gensim import models, test
from gensim.test import utils
from gensim.test.utils import datapath
import numpy as np
import csv

drive.mount('/content/drive/')
os.chdir('/content/drive/Shareddrives/CS260-Project/data/')
```

```python
lda_model = models.ldamodel.LdaModel.load("/content/drive/Shareddrives/CS260-Project/models/lda/big-lda-train-10") #TODO: change file path to folder with lda model

topics = {}
for idx, topic in lda_model.show_topics(formatted=False, num_words=20, num_topics=10):
    # topics[idx] = [w[0] for w in topic]
    topics[idx] = str(idx) + ": "+ " ".join([w[0] for w in topic])


# Modify train data 
artist = []
topic_id = []
lyric = []
with open('./train/big-lda-train-10.csv') as traindata:
  line = 1
  reader = csv.reader(traindata, delimiter=',')
  for row in reader:
    if line > 1:
      artist.append(row[0])
      topic_id.append(row[1])
      lyric.append(row[2])
    line += 1
print(len(set(topic_id)))
topic_id_updated = [topics[int(id)] for id in topic_id]
print(len(set(topic_id_updated)))

with open('./train_updated/big-lda-train-10.csv', 'w') as testdata: #TODO: change file path to new folder
  writer = csv.writer(testdata, delimiter=',')
  writer.writerow(['artist', 'topics', 'lyric'])
  line = 0
  for i in range(len(artist)):
    writer.writerow([artist[i], topic_id_updated[i], lyric[i]])
    line += 1


# Modify test data
artist = []
topic_id = []
lyric = []
with open('./test/big-lda-test-10.csv') as traindata:
  line = 1
  reader = csv.reader(traindata, delimiter=',')
  for row in reader:
    if line > 1:
      artist.append(row[0])
      topic_id.append(row[1])
      lyric.append(row[2])
    line += 1

topic_id_updated = [topics[int(id)] for id in topic_id]

with open('./test_updated/big-lda-test-10.csv', 'w') as testdata: #TODO: change file path to new folder
  writer = csv.writer(testdata, delimiter=',')
  writer.writerow(['artist', 'topics', 'lyric'])
  line = 0
  for i in range(len(artist)):
    writer.writerow([artist[i], topic_id_updated[i], lyric[i]])
    line += 1


# Modify val data
artist = []
topic_id = []
lyric = []
with open('./val/big-lda-val-10.csv') as traindata:
  line = 1
  reader = csv.reader(traindata, delimiter=',')
  for row in reader:
    if line > 1:
      artist.append(row[0])
      topic_id.append(row[1])
      lyric.append(row[2])
    line += 1

topic_id_updated = [topics[int(id)] for id in topic_id]

with open('./val_updated/big-lda-val-10.csv', 'w') as testdata: #TODO: change file path to new folder
  writer = csv.writer(testdata, delimiter=',')
  writer.writerow(['artist', 'topics', 'lyric'])
  line = 0
  for i in range(len(artist)):
    writer.writerow([artist[i], topic_id_updated[i], lyric[i]])
    line += 1
```
