```python
# open csv file for lyrics-data.csv
import csv
import pandas as pd

dataset = []

# Dataset 1
lyrics_info = []
with open('./artist-data/lyrics-data.csv') as lyricsfile:
  line = 1
  reader = csv.reader(lyricsfile, delimiter=',')
  for row in reader:
    if line > 1:
      lang = row[4]
      if lang == 'en':
        artist_link = row[0]
        lyrics = row[3]
        lyrics_info.append((artist_link, lyrics))
    line += 1

artist_info = {}
with open('./artist-data/artists-data.csv') as artistsfile:
  line = 1
  reader = csv.reader(artistsfile, delimiter=',')
  for row in reader:
    if line > 1:
      artist_name = row[0]
      artist_link = row[4]
      genres = row[1]
      artist_info[artist_link] = [artist_name, genres]
    line += 1

for artist, lyrics in lyrics_info:
  if artist in artist_info:
    example = artist_info[artist]
    dataset.append([example[0],lyrics])

# Dataset 3
with open('./artist-data-3/tcc_ceds_music.csv') as lyricsfile:
  line = 1
  reader = csv.reader(lyricsfile, delimiter=',')
  for row in reader:
    if line > 1:
      artist_name= row[1]
      genres = row[4]
      lyrics = row[5]
      dataset.append([artist_name, lyrics])
    line += 1

# Dataset 4
missing_lyrics = "lyrics for this song have yet to be released please check back once the song has been released"
lyric_files = os.listdir("./artist-data-4/")
for file in lyric_files:
  if file != 'kaggle-dataset.csv':
    lyric_idx = -1
    artist_idx = -1
    with open('./artist-data-4/' + file) as lyricsfile:
      line = 0
      reader = csv.reader(lyricsfile, delimiter=',')
      for row in reader:
        if line == 0:
          lyric_idx = row.index('Lyric')
          artist_idx = row.index('Artist') 
        if line > 1 and row[lyric_idx] != missing_lyrics:
          artist_name = row[artist_idx]
          lyrics = row[lyric_idx]
          dataset.append([artist_name, lyrics])
        line += 1


print(len(dataset))
print(dataset[0])
```

```python
# Preprocessing step
import re
escape_seq = ['\n', '\t']
for row in dataset:
  row[0] = row[0].lower()
  row[1] = row[1].lower()
  if escape_seq[0] in row[1] or escape_seq[1] in row[1]:
    row[1] = re.sub(escape_seq[0],' ', row[1])
    row[1] = re.sub(escape_seq[1],' ', row[1])

print(dataset[0])
```

```python
#STEP FOUR: Write this information to a CSV file
with open('./big-kaggle-dataset.csv', 'w') as datasetfile:
  writer = csv.writer(datasetfile)
  writer.writerow(['artist', 'lyrics'])
  for example in dataset:
    writer.writerow(example)
```
