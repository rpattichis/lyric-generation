```python
# open csv file for lyrics-data.csv
import csv
import pandas as pd

# STEP ONE: process the artist to lyric link
lyrics_info = []
with open('./artist-data/lyrics-data.csv') as lyricsfile:
  line = 1
  reader = csv.reader(lyricsfile, delimiter=',')
  for row in reader:
    if line > 1:
      lang = row[4]
      if lang == 'en':
        artist_link = row[0]
        lyrics = row[3]
        lyrics_info.append((artist_link, lyrics))
    line += 1

print(len(lyrics_info))

# STEP TWO: process the artist to genres link
artist_info = {}
with open('./artist-data/artists-data.csv') as artistsfile:
  line = 1
  reader = csv.reader(artistsfile, delimiter=',')
  for row in reader:
    if line > 1:
      artist_name = row[0]
      artist_link = row[4]
      genres = row[1]
      artist_info[artist_link] = [artist_name, genres]
    line += 1
```

```python
# STEP THREE: Map the genres to the lyrics using the artist links
dataset = [] # this will be x -> y, where x is genre and y is the lyrics outputted
for artist, lyrics in lyrics_info:
  if artist in artist_info:
    example = artist_info[artist]
    dataset.append([example[0], example[1], lyrics])

print(len(dataset))
print(dataset[150])
```

```python
#STEP FOUR: Write this information to a CSV file
with open('./artist-data/kaggle-dataset.csv', 'w') as datasetfile:
  writer = csv.writer(datasetfile)
  writer.writerow(['artist', 'genres', 'lyrics'])
  for example in dataset:
    writer.writerow(example)
```
